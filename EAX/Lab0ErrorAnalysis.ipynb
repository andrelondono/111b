# 4) Exponential 

```python
rand_num=[]
N=100
for i in np.arange(N):
    rand_num.append(np.random.normal(0,1))
def func(x):
    return  np.exp(-x)
y = func(np.array(rand_num))
print "Mean:", mean(y)
print "Std: ", std(y)
plt.figure()
plt.title("N={}".format(N))
plt.hist(y,bins=60)
plt.show()
```

```python
N_list = [10,50,100,1000]
for N in N_list:
    rand_num=[]
    mean_list=[]
    for m in np.arange(1000):
        for i in np.arange(N):
            rand_num.append(np.random.normal(0,1))
        def func(x):
            return  np.exp(-x)
        y = func(np.array(rand_num))
        mean_list.append(mean(y))
    plt.figure()
    plt.hist(mean_list,bins=50)
    plt.title("Histograms of Means N={}".format(N))
    plt.xlabel("mean",fontsize=13)
    plt.ylabel("count",fontsize=13)

    print "N={}".format(N)
    print "mean :{}".format(mean(mean_list))
    print "standard deviation: {}".format(std(mean_list))
    print "Number of experiments within 1 sigma: ", len(np.where(abs(mean_list-mean(mean_list))<std(mean_list))[0])
    print "Number of experiments within 2 sigma: ", len(np.where(abs(mean_list-mean(mean_list))<2*std(mean_list))[0])
    print "Number of experiments within 3 sigma: ", len(np.where(abs(mean_list-mean(mean_list))<3*std(mean_list))[0])
```
